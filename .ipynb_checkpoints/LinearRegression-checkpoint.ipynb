```python
x_train= data['TV'].values.reshape(-1, 1)  # Feature
y_train= data['Sales'].values.reshape(-1, 1)  # Target

print(x_train.shape, y_train.shape)
```

```python
print ('The shape of x_train is:', x_train.shape)
print ('The shape of y_train is: ', y_train.shape)
print ('Number of training examples (m):', len(x_train))
```

```python
n=len(x_train)
```

```python
#min-max Scaling
x_min=np.min(x_train)
x_max=np.max(x_train)
y_min=np.min(y_train)
y_max=np.max(y_train)
x_train=(x_train-x_min)/(x_max-x_min)
y_train=(y_train-y_min)/(y_max-y_min)


```

```python
def compute_cost(x, y, w, b):
      
    total_cost = 0.0  # Initialize cost
    # Compute squared errors
    for i in range(n):
        total_cost += ((w * x[i] + b) - y[i]) ** 2  
    return total_cost / (2 * n) # Ensure a Python float is returned

def vec_compute_cost(x,y,w,b):

    total_cost=0
   
    total_cost=np.sum(((w*x+b)-y)**2)
    
    return (total_cost)/(2*n)


    
```

```python
initial_w=2
initial_b=1
cost1=compute_cost(x_train,y_train,initial_w,initial_b)
cost2=vec_compute_cost(x_train,y_train,initial_w,initial_b)

print(type(cost2))
print(f"Initial Cost (Before Training): {cost2:.3f}")
```

```python
b_values = np.linspace(-50, 50, 100)  # Varying b from -50 to 50
w_fixed = 2  # Keeping w constant (can choose any value)


cost_values = [compute_cost(x_train, y_train, w_fixed, b) for b in b_values]

# Plot cost function
plt.figure(figsize=(8, 5))
plt.plot(b_values, cost_values, color='r', linewidth=2)

# Add labels and title
plt.xlabel("Bias (b)", fontsize=14)
plt.ylabel("Cost J(b)", fontsize=14)
plt.title("Cost Function vs. Bias", fontsize=16)
plt.grid(True)

# Show plot
plt.show()
```

```python
def compute_gradient(x,y,w,b):
    m=x_train.shape[0]
    df=np.sum((w*x+b-y)*x)/m
    db=np.sum((w*x+b-y))/m
    return df,db

    
```

```python
initial_w = 0
initial_b = 0
tmp_dj_dw, tmp_dj_db = compute_gradient(x_train, y_train, initial_w, initial_b)
print('Gradient at initial w, b (zeros):', tmp_dj_dw, tmp_dj_db)
```

```python
import copy
import math

def gradient_descent(x, y, w_in, b_in, cost_function, gradient_function, alpha, num_iters): 

    x = x.flatten()
    y = y.flatten()

    # Deep copy to avoid modifying global variables
    w = copy.deepcopy(w_in)
    b = b_in

    J_history = []  # Stores cost function values
    w_history = []  # Stores weight updates for visualization

    for i in range(num_iters):
        # Compute gradients
        dj_dw, dj_db = gradient_function(x, y, w, b)

        # Update parameters
        w -= alpha * dj_dw
        b -= alpha * dj_db

        # Compute cost and store history
        cost = vec_compute_cost(x, y, w, b)
        J_history.append(cost)

        # Store weights every few iterations for analysis
        if i % math.ceil(num_iters / 10) == 0:
            w_history.append(w)
            print(f"Iteration {i:4}: Cost = {cost:.4f}, w = {w:.4f}, b = {b:.4f}")

    return w, b, J_history, w_history
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate w and b values for the contour plot
w_values = np.linspace(-10, 10, 100)  # Range for w
b_values = np.linspace(-10, 10, 100)  # Range for b
W, B = np.meshgrid(w_values, b_values)  # Create grid

# Compute cost for each (w, b)
Costs = np.zeros(W.shape)
for i in range(W.shape[0]):
    for j in range(W.shape[1]):
        Costs[i, j] = vec_compute_cost(x_train, y_train, W[i, j], B[i, j])  

# Run gradient descent to get history of w, b values
w_final, b_final, J_history, w_history = gradient_descent(x_train, y_train, initial_w, initial_b, 
                                                           vec_compute_cost, compute_gradient, 
                                                           alpha=0.01, num_iters=50)
# Extract w, b history for plotting
w_hist = np.array(w_history)
b_hist = np.array([b_final for _ in range(len(w_hist))])  # Assuming b is stored at last step

# Plot contour
plt.figure(figsize=(8, 6))
plt.contour(W, B, Costs, levels=50, cmap='viridis')  # Contour plot
plt.colorbar(label='Cost')  # Color bar
plt.plot(w_hist, b_hist, 'r.-', markersize=5, label='Gradient Descent Path')

# Labels and title
plt.xlabel("Weight (w)", fontsize=12)
plt.ylabel("Bias (b)", fontsize=12)
plt.title("Cost Contour with Gradient Descent Path", fontsize=14)
plt.legend()
plt.grid(True)

plt.show()
```

```python
def linear_regression(X_train, y_train, w, b, alpha, num_iters):
    w_final, b_final, cost_history, w_history = gradient_descent(
    X_train, y_train, w, b, compute_cost, compute_gradient, alpha, num_iters)

    # Make predictions
    y_pred = X_train * w_final + b_final
    
    # Evaluate performance
    

    return w_final, b_final, num_iters, cost_history[-1]
```
